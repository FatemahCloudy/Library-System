```python
#run first before any other function

def cleaning(file):
    lis = []
    #each line
    for i in file:
        i = i.strip()
        #each word
        for k in i:
            #each char
            for q in k:
                if q in "''()" :
                    i= i.replace(q,"")
        #each item
        i = i.split(",")
        for l in i:
            m = l.strip()
            i[i.index(l)] = m
        j = tuple(i)
        lis.append(j)
    file.close()
    return lis
```

```python
#displaying the list of books

def bookList():
    #open the file
    file = open('books.txt')
    #return a list of books as tuples
    return cleaning(file)
```

```python
#displaying the list of borrowers
def borrowerList():
    #opinning the file
    file = open('borrowers.txt')
    #returnung a list of borrowers as tuples
    return cleaning(file)
```

```python
#finding a borrowers by their ID number
def searchID (idn):
    flag = False
    #getting the list of borrowers
    lis = borrowerList()
    #iterating through each tuple
    for i in lis:
        #checking if the IDs mach
        if i[0] == idn:
            flag =  True
            #returning the borrower informations
            return tuple(i)
    return flag
```

```python
#finding a book by its ISPN

def searchISBN (isbn):
    flag = False
    #getting the list of books
    lis = bookList()
    #iterating through each tuple
    for i in lis:
        #checking if the ISPNs mach
        if i[0] == isbn:
            flag =  True
            #returning the book informations
            return tuple(i)   
    return flag
```

```python
#adding a new borrower to the list
def addBorrower():
    #get input as id number
    idn = input("Enter the id: ")
    #checking if it exists
    if not searchID(idn):
        #if not, get the other information
        name = input("Enter borrower name: ")
        contact = input("Enter contact e-mail:")
        #oppen the borrowers file in append mode
        with open("borrowers.txt","a") as file:
            #create a string with the new borrower information
            x = f'{idn}, {name}, {contact}\n'
            #write it in the end of the file and close the file
            file.write(x)
        print('Done! The new borrower has been added')
    else:
        print("user already exist")
```

```python
# add a book
def addBook():
    #get input as ISBN
    isbn = input("Enter the isbn: ")
    #checking if it exists
    if not searchISBN(isbn):
        #if not, get the other information
        title = input('Please enter the title: ')
        author = input('Please enter the author name:')
        year = input('Please enter the publication year:')
        availability = input('Please enter the availability status:')
        #oppen the books file in append mode
        with open("books.txt","a") as file:
            #create a string with the new book information
            x = f"{isbn}, {title}, {author}, {year}, {availability}\n"
            #write it in the end of the file and close the file
            file.write(x)
        print('Done! The book has been added')
        
    else:
        print('Book already exist')
```

```python
#getting the updated information of a book with the original in a list

def update():
    #gitting the input as ISPN
    isbn = input("Enter the ISBN: ")
    #checking if it exists
    book = searchISBN(isbn)
    #in case it doesn't
    if not book:
        print('This book does not exist')
        return False
    #in case it exists
    else:
        #getting the book information in a list instead of a tuple
        result = list(book)
        # asking what item does the user want to change
        editted = int(input("type 1 for name, 2 for auther, 3 for year, or 4 for status: "))
        #change book title
        if editted == 1:
            result[1] = input("Enter updated title: ")
        #change auther name
        elif editted == 2:
            result[2] = input("Enter updated auther: ")
        # change publication year
        elif editted == 3:
            result[3] = (input("Enter updated publication year: "))
        elif editted == 4:
            result[4] = (input("Enter updated status: "))
        #return a list of the old information tuple and the new information tuple
        return [book, tuple(result)]
    
#update the book information in the file
def newBooks():
    #open the books file
    filer = open("books.txt", 'r')
    # calling the function of new information
    new = update()
    #if the book exists
    if new:
        k = ""
        # iterating through the lines of the file(each line is a tuple for a book)
        for line in filer:
            #adding the line to a string
            k +=line
        #replacing the old information with the new one
        file1 = k.replace(str(new[0]), str(new[1]).strip())
        #opening the file in write mood (clear any data)
        file = open("books.txt", 'w')
        #writing the whole updated text
        file.write(file1) 
    #closing both files
    file.close()
    filer.close()
    print('Information has been updated successfully')
```

```python
#removing a book information from the file

def deleteBook():
    #get input as id number
    isbn = input("Enter the ISBN: ")
    #checking if it exists
    book = searchISBN(isbn)
    #if it exists get confirmation from the user
    if book:
        flag = input("Are you sure you want to delete? type Y for yes or N for no: ")
        #if the user confirmed open the file in read mood
        if flag.lower() =='y':
            filer = open("books.txt", 'r')
            k = ""
            # iterating through the lines of the file(each line is a tuple for a book)
            for line in filer:
                #adding the line to a string
                k +=line
                #replacing the book information with the new line by an empty string
            file1 = k.replace(str(book)+"\n","")
                #opening the file in write mood (clear any data)
            file = open("books.txt", 'w')
                #writing the whole updated text
            file.write(file1.strip()) 
            #close both files
            file.close()
            filer.close()
            print(f'Done! {book[1]} is deleted')
        else:
            print('Deleting is canceled')
```

```python
def updateStatus(isbn, status):
    #checking if it exists
    book = searchISBN(isbn)
    #in case it doesn't
    if not book:
        print('This book does not exist')
        return False
    #in case it exists
    else:
        #getting the book information in a list instead of a tuple
        result = list(book)
        result[4] = status
        new = [book, tuple(result)]
        books = open ('books.txt')
        lines = ""
        # iterating through the lines of the file(each line is a tuple for a book)
        for line in books:
            #adding the line to a string
            lines +=line
        #replacing the old information with the new one
        update = lines.replace(str(new[0]), str(new[1]).strip())
        #opening the file in write mood (clear any data)
        file = open("books.txt", 'w')
        #writing the whole updated text
        file.write(update) 
        #closing both files
        file.close()
        books.close()
```
